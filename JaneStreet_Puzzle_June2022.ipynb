```python
# find positions within puzzle that are empty / that need to be filled 
def find_empty(board, boxes):
  for box in boxes:
    for i in range(len(board)):
        for j in range(len(board[0])):
            if (box[i][j] == 1) & (board[i][j] == 0): 
                # print(i,j)
                return (i, j) # row, column
  return None
```

```python
def valid(board, num, pos):
    #find the box it belongs to
    for i in boxes:
        if i[pos] == 1: 
            if num > sum(i.reshape(-1)):
                # print(str(pos) + " number too large: "+ str(num))
                return False
            box_board = i*board
            if num in box_board.reshape(-1):
                # print(str(pos) + " number already in box: "+ str(num))
                return False
            else: 
                min_dist = 100
                for row in range(len(board)): 
                    for column in range(len(board)): 
                        if board[row][column] == num:
                            min_dist = min(taxicap_distance(pos, (row, column)), min_dist)
                            if min_dist < num: 
                                # print(str(pos) + " incorrect distance: "+ str(num))
                                return False
    # print(str(pos) + " possible solution: " + str(num))
    return True
```

```python
def taxicap_distance(pos1, pos2):
    return sum(abs(val1-val2) for val1, val2 in zip (pos1, pos2))
```

```python
invalid_pos = []
```

```python
def solve(start_board, board, boxes):
    #creat list for positions with incorrect numbers
    invalid_pos = []
      
    #find the next empty field in the box
    find = find_empty(board, boxes)
    print(find)
    
    #if board is full, check whether current solution is correct
    if not find:
        print(start_board)
        
        
        #loop through all numbers to check distances
        for num2 in range(1,11):
            pos_list = []
            #first find positions that have number num2
            for row in range(len(board)): 
                    for column in range(len(board)): 
                        if board[row][column] == num2:
                            pos_list.append((row,column))
            if len(pos_list) != 0:
                for idx in range(len(pos_list)):
                    min_dist = 100
                    for idx2 in range(len(pos_list)):
                        if idx != idx2: 
                            min_dist = min(taxicap_distance(pos_list[idx], pos_list[idx2]), min_dist)
                    if min_dist != num2:
                        invalid_pos.append(pos_list[idx])
        if len(invalid_pos) != 0:
            board = start_board
            print(board)
            find = find_empty(board, boxes)
            row, column = find
            return False
        else: 
            return True
    else: 
        row, column = find
    
    for num in range(1, 11):
        if (row, column) in invalid_pos: 
            num = num + 1
        if valid(board, num, (row, column)):
            board[row][column] = num
            
            if solve(start_board, board, boxes):
                return True
            
            board[row, column] = 0
    
    return False
```
